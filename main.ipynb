## Part 2

This part consist of defining a function that calculate the distance between 2 geographical points: It taskes the longitude and latitude as inputs:

```python
from math import sin, cos, sqrt, atan2
def distance(lat1,lon1,lat2,lon2):
    R = 6373.0
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance
```

Next we will need to extract the long and lat given an ID:

```python
## df_2000.loc[(df['id']==7397300170),'long'].values[0]
def distance_id(ID1,ID2,df):
    lat1=df.loc[(df['id']==ID1),'lat'].values[0]
    long1=df.loc[(df['id']==ID1),'long'].values[0]
    lat2=df.loc[(df['id']==ID2),'lat'].values[0]
    long2=df.loc[(df['id']==ID2),'long'].values[0]
    return distance(lat1,long1,lat2,long2)
    


    
```
