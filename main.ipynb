```python
df = pd.read_csv('kc_house_data.csv')
```

We will start by taking the houses built from 2000 onwards. We will put it in a seperate dataFrame named df_2000.

```python
df_2000=df.loc[df['yr_built']>=2000].copy()
```

We will check how many houses we have left:

```python
df_2000.shape
```

So there is 4759 houses left.

We need to find the most expensive house for each year. We will put all the results in a new dataFrame named df_exp.
According to that, we will start by putting all the year values in a separate list:
Note that we should remove the duplicates by applying the set function.

```python
yr_built=list(set(df_2000['yr_built'].values))
```

Now we will create an empty dataFrame and concat the row of the most expensive house for each year:

```python
df_exp=pd.DataFrame()
```

```python
for y in yr_built:
    df_exp=pd.concat([df_exp, df_2000.loc[df_2000['yr_built']==y].nlargest(1,'price')], axis=0, ignore_index=True)
```

```python
list_columns=['id','price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','condition','grade','yr_built','waterfront','view'] #list of column to display
```

```python
df_exp[list_columns]
```

We need to find the houses for the most bedroom, bathroom and floors respectively. We will put them in seperate dataFrames named respectively df_maxbd , df_maxbth and df_maxflr

```python
df_maxbd=pd.DataFrame()
df_maxbth=pd.DataFrame()
df_maxflr=pd.DataFrame()
```

```python
for y in yr_built:
    df_maxbd=pd.concat([df_maxbd, df_2000.loc[df_2000['yr_built']==y].nlargest(1,'bedrooms')], axis=0, ignore_index=True)
    df_maxbth=pd.concat([df_maxbth, df_2000.loc[df_2000['yr_built']==y].nlargest(1,'bathrooms')], axis=0, ignore_index=True)
    df_maxflr=pd.concat([df_maxflr, df_2000.loc[df_2000['yr_built']==y].nlargest(1,'floors')], axis=0, ignore_index=True)
```
